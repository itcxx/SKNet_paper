```python
import numpy as np
import cv2
import torch
import matplotlib.pyplot as plt

def unnormalize(img_tensor, mean, std):
    """
    将 (C,H,W) 的图像张量反归一化到 [0,255] 范围。
    mean, std 通常是形如 (3,) 的 RGB 均值和标准差，且已被 /255 处理过。
    """
    # (C,H,W) -> numpy
    img = img_tensor.cpu().numpy()
    # 通道级反归一化
    for c in range(img.shape[0]):
        img[c] = img[c] * std[c] + mean[c]
    img = np.clip(img, 0, 1)  # 这里恢复到 [0,1]
    img = (img * 255.0).astype(np.uint8)
    # 转成 (H,W,C)
    img = np.transpose(img, (1, 2, 0))
    return img

def draw_2d_keypoints(image, keypoints_2d):
    """
    在图像上绘制关键点 (K, 3): (x, y, visible)。
    image: (H,W,3)的uint8图像 (BGR或RGB都可以，但后面显示注意颜色)
    """
    img_drawn = image.copy()
    for (x, y, v) in keypoints_2d:
        if v > 0:
            cv2.circle(img_drawn, (int(x), int(y)), 4, (0, 255, 0), -1)
    return img_drawn

def visualize_3d_heatmaps_on_2d(image, heatmaps_3d):
    """
    将 3D 热图 (K, D, H, W) 在深度维度上取最大，得到 (K, H, W)；
    再对 K 个通道做均值 -> (H, W)，最后叠加伪彩色到图像上。
    """
    K, D, H, W = heatmaps_3d.shape
    # 在深度维度上取最大
    heatmaps_2d = np.max(heatmaps_3d, axis=1)  # (K, H, W)
    # 对 K 个通道取均值 -> (H, W)
    heatmap_2d = np.mean(heatmaps_2d, axis=0)
    # 归一化
    heatmap_2d = (heatmap_2d - heatmap_2d.min()) / (heatmap_2d.max() - heatmap_2d.min() + 1e-5)
    heatmap_2d = (heatmap_2d * 255).astype(np.uint8)
    heatmap_color = cv2.applyColorMap(heatmap_2d, cv2.COLORMAP_JET)
    # 调整大小到原图维度
    heatmap_color = cv2.resize(heatmap_color, (image.shape[1], image.shape[0]))
    # 叠加
    overlay = cv2.addWeighted(image, 0.6, heatmap_color, 0.4, 0)
    return overlay
```

```python
%matplotlib inline
import yaml
import torch
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

# 根据你的实际项目路径导入
from datasets import HandPoseDataset
from utils.build_pipeline import build_pipeline
from utils.collate import custom_collate_fn

def inspect_batch(config_path="./config/config.yaml", batch_size=4):
    """
    读取config，构建数据集，并一次取出一个batch，展示所有样本的可视化结果
    """
    # 1. 读取配置文件
    with open(config_path, "r") as f:
        config = yaml.safe_load(f)
    
    # 2. 构建 pipeline (这里假设train_pipeline存在)
    transform = build_pipeline(config, pipeline_key="train_pipeline")

    # 3. 构建 dataset & dataloader
    dataset = HandPoseDataset(config_path=config_path, mode="train", transform=transform)
    dataloader = DataLoader(
        dataset,
        batch_size=batch_size,
        shuffle=True,
        num_workers=0,
        collate_fn=custom_collate_fn
    )

    # 4. 取第一个 batch
    batch = next(iter(dataloader))

    # 取出图像张量 (B,C,H,W) 和 关键点 (B, K, 3) 以及 3D 热图 (B, K, D, H, W)
    imgs = batch["img"]                # shape (B, C, H, W)
    kpts_2d = batch["keypoints_2d"]    # shape (B, K, 3)
    meta_info = batch["meta_info"]
    heatmaps_3d = meta_info.get("heatmaps", None)  # shape (B, K, D, H, W) if it exists

    # 从 config 里读取用于反归一化的 mean/std
    mean = config["data_preprocessor"]["mean"]  # [123.675, 116.28, 103.53] 之类
    std = config["data_preprocessor"]["std"]    # [58.395, 57.12, 57.375] 之类

    # 先把它们除以255，保证跟 transforms.Normalize() 的用法对应
    mean = [m / 255.0 for m in mean]
    std = [s / 255.0 for s in std]

    B = imgs.shape[0]

    # 5. 可视化每个样本
    #   - 若 batch 很大，可以只看前几张
    for i in range(B):
        # (C,H,W) -> 反归一化 -> (H,W,C)
        img_np = unnormalize(imgs[i], mean, std)

        # 关键点
        kpt2d = kpts_2d[i].cpu().numpy()  # (K,3)

        # 在图上画关键点
        img_with_kpts = draw_2d_keypoints(img_np, kpt2d)

        # 如果有 3D heatmaps
        if heatmaps_3d is not None:
            hm_3d = heatmaps_3d[i].cpu().numpy()  # (K,D,H,W)
            img_with_hm = visualize_3d_heatmaps_on_2d(img_with_kpts, hm_3d)
        else:
            img_with_hm = img_with_kpts

        # 用 matplotlib 展示
        plt.figure(figsize=(6,6))
        plt.title(f"Sample {i} in batch")
        plt.imshow(cv2.cvtColor(img_with_hm, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.show()

# 调用示例
inspect_batch(config_path="./config/config.yaml", batch_size=4)
```
